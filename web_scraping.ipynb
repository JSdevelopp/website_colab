```python
import requests
from bs4 import BeautifulSoup
import re
import random
```

```python
base_url = "http://books.toscrape.com/catalogue/page-1.html"
```

```python
list_of_images = []
for page in range(1, 50):
    base_url = f"http://books.toscrape.com/catalogue/page-{page}.html"
    response = requests.get(base_url)
    t = response.text
    soup = BeautifulSoup(t, features = 'html.parser')
    for i in range(len(soup.find_all('a'))):
        if 'src' in str(soup.find_all('a')[i]):  
            pattern = re.escape('src') + r'.+'
            the_src = re.findall(pattern , str(soup.find_all('a')[i]))[0]
            pattern2 = r'"(.*?)"'
            list_of_images.append('http://books.toscrape.com/' + re.findall(pattern2, the_src)[0])
            
```

```python
list_of_prices = []

for page in range(1, 50):
    base_url = f"http://books.toscrape.com/catalogue/page-{page}.html"
    response = requests.get(base_url)
    t = response.text
    soup = BeautifulSoup(t, features = 'html.parser')
    for i in soup.find_all(class_="price_color"):
        pattern2 = r'>(.*?)<'
        pattern3 = r'/d+'
        


        list_of_prices.append(float(re.findall(pattern2, str(i))[0][2:]))
        
          
```
